```python
# Import required libraries
import pandas as pd
import numpy as np

# Load the Spotify dataset into DataFrame
df = pd.read_csv("SpotifY_DataSet1.csv")

# Display first 5 rows of the dataset
print("First 5 Rows:")
print(df.head())

# Display dataset structure and data types
print("\nDataset Information:")
print(df.info())

# Display statistical summary of numerical columns
print("\nStatistical Description:")
print(df.describe())

# Check for missing values in each column
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Remove duplicate rows from the dataset
df = df.drop_duplicates()

# Fill missing numeric values with column mean
for col in df.select_dtypes(include=["float64", "int64"]).columns:
    df[col] = df[col].fillna(df[col].mean())

# Fill missing categorical values with 'Unknown'
for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].fillna("Unknown")

# Verify missing values after cleaning
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
import numpy as np

# --------- CONVERT DURATION FROM MILLISECONDS TO MINUTES ----------
# This creates a new derived column 'duration_min'
if "duration_ms" in df.columns:
    df["duration_min"] = df["duration_ms"] / 60000


# --------- CREATE POPULARITY LEVEL COLUMN ----------
# High if popularity > 60, else Low
if "popularity" in df.columns:
    df["popularity_level"] = np.where(df["popularity"] > 60, "High", "Low")

# Step 1: Replace 'Unknown' with 0 (non-explicit)
if "explicit" in df.columns:
    df["explicit"] = df["explicit"].replace("Unknown", 0)

    # Step 2: Convert safely to integer
    df["explicit"] = df["explicit"].astype(int)


# --------- DISPLAY UPDATED DATASET ----------
print("Dataset after Data Wrangling:")
print(df.head())
```

```python
# Filter songs with popularity greater than 70
popular_songs = df[df["popularity"] > 70]
print("Popular Songs:")
print(popular_songs.head())

# Apply multiple conditions using AND operator
dance_energy = df[(df["danceability"] > 0.7) & (df["energy"] > 0.6)]
print("Danceable & Energetic Songs:")
print(dance_energy.head())

# Set first column as index for better data access
df.set_index(df.columns[0], inplace=True)

# Display dataset after indexing
print(df.head())
```

```python
# --------- DISPLAY ALL COLUMN NAMES ----------
# This helps to verify available columns in the dataset
print("Columns in Dataset:")
print(df.columns)


# --------- GROUPING USING A VALID COLUMN ----------
# We safely choose a categorical column that exists
# Example: 'playlist_genre' or 'track_genre' is commonly present in Spotify datasets

group_column = df.select_dtypes(include=["object"]).columns[0]   # first categorical column
numeric_column = df.select_dtypes(include=["int64", "float64"]).columns[0]  # first numeric column

# Group the data using the selected columns
grouped_data = df.groupby(group_column)[numeric_column].mean()

# Display grouped result
print("\nGrouped Data (Mean Calculation):")
print(grouped_data)


# --------- APPLY MULTIPLE AGGREGATION FUNCTIONS ----------
# Applying sum, mean, count, min, median, std on numeric columns
aggregation_result = df.select_dtypes(include=["int64", "float64"]).agg(
    ["sum", "mean", "count", "min", "median", "std"]
)

# Display aggregation table
print("\nAggregation Results:")
print(aggregation_result)
```

```python
import matplotlib.pyplot as plt

# --------- DISPLAY COLUMN NAMES ----------
# This helps confirm which columns are available
print("Available Columns:")
print(df.columns)


# --------- SELECT VALID NUMERIC & CATEGORICAL COLUMNS ----------
categorical_col = df.select_dtypes(include=["object"]).columns[0]
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

x_col = categorical_col          # X-axis column
y_col = numeric_cols[0]          # Y-axis column for plotting


# --------- 1. LINE PLOT ----------
grouped_line = df.groupby(x_col)[y_col].mean()

plt.figure()
grouped_line.plot()
plt.title("Line Plot of " + y_col + " grouped by " + x_col)
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.show()


# --------- 2. SCATTER PLOT ----------
# Use first two numeric columns
if len(numeric_cols) >= 2:
    plt.figure()
    plt.scatter(df[numeric_cols[0]], df[numeric_cols[1]])
    plt.title("Scatter Plot between Numeric Columns")
    plt.xlabel(numeric_cols[0])
    plt.ylabel(numeric_cols[1])
    plt.show()


# --------- 3. BAR CHART ----------
bar_data = df[x_col].value_counts().head(10)

plt.figure()
bar_data.plot(kind="bar")
plt.title("Bar Chart of " + x_col)
plt.xlabel(x_col)
plt.ylabel("Count")
plt.show()


# --------- 4. HISTOGRAM ----------
plt.figure()
plt.hist(df[y_col], bins=10)
plt.title("Histogram of " + y_col)
plt.xlabel(y_col)
plt.ylabel("Frequency")
plt.show()
```
